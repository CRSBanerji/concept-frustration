# Sarcasm task

This notebook mirrors the structure of the synthetic and globe notebooks:
1. set parameters
2. run the experiment
3. save and summarise results


```
import sys
from pathlib import Path

import pandas as pd
import torch

repo_root = Path.cwd().resolve().parent
src_path = repo_root / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

from sarcasm_task.config import RESULTS_DIR
from sarcasm_task.experiment import run_experiment, save_results, summarize_results
```

```
# ----------------------------
# Run configuration
# ----------------------------
device = "cuda" if torch.cuda.is_available() else "cpu"

# dataset / folds
N_DATA = 10000
SEED_DATA = 123
K_FOLDS = 10
SEED_FOLDS = 777

# Fisher window
P_LO = 0.2
P_HI = 0.8
MIN_KEEP = 500

# model / training
HIDDEN_BB = 128
EPOCHS_BB = 60
EPOCHS_CBM = 30
LAMBDA_C = 1.0
K_SAE = 300
EPOCHS_SAE = 60

# preprocessing toggles
STANDARDIZE_X_PER_FOLD = True
NORMALIZE_C_FOR_CBM = True
NORM_EPS = 1e-6

# frustration metric
PAIR_SOFT_EPS = 1e-6

# output
RESULTS_NAME = "results_sarcasm.csv" 
```

```
rows = run_experiment(
    n_data=N_DATA,
    seed_data=SEED_DATA,
    k_folds=K_FOLDS,
    seed_folds=SEED_FOLDS,
    device=device,
    p_lo=P_LO,
    p_hi=P_HI,
    min_keep=MIN_KEEP,
    hidden_bb=HIDDEN_BB,
    epochs_bb=EPOCHS_BB,
    epochs_cbm=EPOCHS_CBM,
    lambda_c=LAMBDA_C,
    K_sae=K_SAE,
    epochs_sae=EPOCHS_SAE,
    standardize_x_per_fold=STANDARDIZE_X_PER_FOLD,
    normalize_c_for_cbm=NORMALIZE_C_FOR_CBM,
    norm_eps=NORM_EPS,
    pair_soft_eps=PAIR_SOFT_EPS,
)
```

```
out_path = RESULTS_DIR / RESULTS_NAME
df = save_results(rows, out_path)
summary = summarize_results(rows)[1]

display(df.head())
display(summary)
```
