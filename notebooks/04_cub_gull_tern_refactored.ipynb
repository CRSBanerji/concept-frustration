# CUB Gull vs Tern concept-frustration experiment

Refactored notebook with editable parameter block, matching the globe and synthetic notebooks.

```python
from pathlib import Path
import sys
import pandas as pd
import torch

repo_root = Path.cwd().resolve().parent
sys.path.append(str(repo_root / "src"))

from cub_task.config import RESULTS_DIR, default_results_path
from cub_task.experiment import run_experiment
```

```python
# ----------------------------
# Run configuration
# ----------------------------
device = "cuda" if torch.cuda.is_available() else "cpu"

# folds / seeds
K_FOLDS = 10
SEED_FOLDS = 777

# Fisher window
P_LO = 0.2
P_HI = 0.8
MIN_KEEP = 200

# models
K_SAE = 300
PAIR_SOFT_EPS = 1e-6
RIDGE_LAM_C3 = 1e-3

# output
RESULTS_PATH = default_results_path(k_folds=K_FOLDS)

print("Using device:", device)
print("Results path:", RESULTS_PATH)
```

```python
df = run_experiment(
    device=device,
    k_folds=K_FOLDS,
    seed_folds=SEED_FOLDS,
    p_lo=P_LO,
    p_hi=P_HI,
    min_keep=MIN_KEEP,
    K_sae=K_SAE,
    pair_soft_eps=PAIR_SOFT_EPS,
    results_path=RESULTS_PATH,
)
```
