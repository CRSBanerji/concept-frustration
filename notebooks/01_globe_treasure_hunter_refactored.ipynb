# Globe / Treasure Hunter experiment

This notebook is a lightweight runner for the refactored package in `src/globe_treasure_hunter/`.
The function bodies are unchanged from the original notebook; they have only been moved into modules.

```
import sys
from pathlib import Path

repo_root = Path.cwd().resolve().parent
if str(repo_root / 'src') not in sys.path:
    sys.path.append(str(repo_root / 'src'))

import numpy as np
import pandas as pd
import torch

from globe_treasure_hunter.experiment import run_sweep, summarize
from globe_treasure_hunter.stats import paired_wilcoxon_report
```

```
device = "cuda" if torch.cuda.is_available() else "cpu"

seeds = np.arange(1, 51)
scenarios = [0, 1]

n = 8000
r = 64
sigma_x = 0.3

p_lo, p_hi = 0.2, 0.8
min_keep = 200

K_sae = 60
pair_soft_eps = 1e-6

d_max = 1.0
R_task = 0.75
```

```
rows = run_sweep(
    scenarios=scenarios,
    seeds=seeds,
    device=device,
    sigma_x=sigma_x,
    p_lo=p_lo,
    p_hi=p_hi,
    min_keep=min_keep,
    r=r,
    n=n,
    K_sae=K_sae,
    pair_soft_eps=pair_soft_eps,
    d_max=d_max,
    R_task=R_task,
)

summarize(rows)
```

```
paired_wilcoxon_report(rows, metrics=("bb_acc", "cbm_acc", "F_pair_raw_mean", "E_pair_raw_mean"))
```

```
df = pd.DataFrame(rows)
df.to_csv("results_globe.csv", index=False)
print("Saved raw results to results_globe.csv")
```
